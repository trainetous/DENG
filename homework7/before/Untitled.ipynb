```python
import graphviz
```

```python
dot = graphviz.Digraph(comment='Security Architecture')
dot.attr(rankdir='TB', splines='ortho')
```

```python
# Define node styles
dot.attr('node', shape='box', style='rounded,filled', fontname='Arial', fontsize='12')

# Add nodes with colors
nodes = [
    ('internet', 'Internet', 'lightcoral'),
    ('api_gw', 'API Gateway/Reverse Proxy\\n(SSL, WAF)', 'lightblue'), 
    ('firewall', 'Firewall/Security Group\\n(Filter Traffic)', 'lightblue'),
    ('orchestrator', 'Container Orchestration\\n(Docker/Kubernetes)', 'lightgreen'),
    ('app', 'Application Container\\n(Non-root, Env Vars)', 'lightgreen'),
    ('database', 'Database/Services\\n(Protected Network)', 'mediumpurple')
]

for node_id, label, color in nodes:
    dot.node(node_id, label, fillcolor=color)

# Add edges
edges = [
    ('internet', 'api_gw'),
    ('api_gw', 'firewall'),
    ('firewall', 'orchestrator'),
    ('orchestrator', 'app'),
    ('app', 'database')
]

for src, dst in edges:
    dot.edge(src, dst)

# Render diagram
dot.render('security_architecture', format='png', cleanup=True)
```
